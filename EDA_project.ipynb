```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Histograms of Numerical Features

```python
numerical_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 2, i + 1)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Count Plots of Categorical Features

```python
categorical_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']

plt.figure(figsize=(18, 15))
for i, col in enumerate(categorical_cols):
    plt.subplot(4, 2, i + 1) # Adjust subplot grid as needed
    sns.countplot(data=df, x=col, hue=col, palette='viridis', legend=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout()
plt.show()
```

## Visualizing Missing Values

```python
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0] # Filter to show only columns with missing values

if not missing_values.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing_values.index, y=missing_values.values, palette='viridis')
    plt.title('Count of Missing Values per Column')
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print('No missing values found in the DataFrame.')
```

## Pie Chart for Loan Status

```python
loan_status_counts = df['Loan_Status'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(loan_status_counts, labels=loan_status_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Distribution of Loan Status')
plt.ylabel('') # Hide the y-label as it's not relevant for pie charts
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Average Loan Amount by Education Level

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Education', y='LoanAmount', data=df, hue='Education', palette='viridis')
plt.title('Average Loan Amount by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Loan Amount')
plt.tight_layout()
plt.show()
```

```python
average_income_by_loan_status = df.groupby('Loan_Status')['ApplicantIncome'].mean()
print('Average ApplicantIncome grouped by Loan_Status:')
print(average_income_by_loan_status)
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='Education', y='LoanAmount', data=df, hue='Education', palette='viridis', legend=False)
plt.title('LoanAmount Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Loan Amount')
plt.tight_layout()
plt.show()
```
